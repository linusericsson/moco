```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='ticks', context='talk')
```

```python
#mAP = [32.7, 37.9, 41.9, 42.9, 40.9, 42.0, 43.0, 43.8]
#FP = [29.31, 15.65, 13.48, 14.09, 14.92, 15.51, 13.27, 11.93]
#FN = [8.38, 15.34, 16.72, 15.22, 15.33, 13.74, 16.18, 16.62]
#miss = [2.53, 7.01, 7.55, 6.54, 5.24, 5.22, 7.47, 7.43]

mAP = np.array([41.9, 42.9, 40.9, 42.0, 43.0, 43.8])
FP = np.array([13.48, 14.09, 14.92, 15.51, 13.27, 11.93])
FN = np.array([16.72, 15.22, 15.33, 13.74, 16.18, 16.62])
miss = np.array([7.55, 6.54, 5.24, 5.22, 7.47, 7.43])

order = np.argsort(mAP)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(14, 4))
axs[0].plot(mAP[order], FP[order])
axs[0].set_ylabel('FP')
axs[0].set_xlabel('mAP')
axs[1].plot(mAP[order], FN[order])
axs[1].set_ylabel('FN')
axs[1].set_xlabel('mAP')
axs[2].plot(mAP[order], miss[order])
axs[2].set_ylabel('Miss')
axs[2].set_xlabel('mAP');

plt.tight_layout()
```
